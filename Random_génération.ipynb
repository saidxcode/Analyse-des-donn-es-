```python
def glc(a, c, m, x_0, N):
  random_numbers = []
  x_n = x_0
  for _ in range(N):
    x_n = (a * x_n + c) % m
    random_numbers.append(x_n / m)
  return random_numbers
  glc()
```

```python
import time

# 1. Define the parameters
m = 2**32
a = 1664525
c = 1013904223

# 2. Set the seed using the current Unix time (converted to an integer)
x_0 = int(time.time())

# 3. Choose how many numbers you want to generate
N = 10

# 4. Call your function
my_random_numbers = glc(a, c, m, x_0, N)

print(my_random_numbers)
```
